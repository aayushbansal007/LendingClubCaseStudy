<h2 style = "color : Brown"> Problem Statement</h2>

<ul>

<h3>Continous Number columns</h3>

```python
#outlier removal functions
def outlier_IQR_method_remove(df):
    A1=df.quantile(0.25)
    A3=df.quantile(0.75)
    IQR=A3-A1
    df_final_lendingcase=df[~((df<(A1-1.5*IQR)) | (df>(A3+1.5*IQR)))]
    return df_final_lendingcase
```

<h4>Column loan_amnt</h4>

```python
#boxplot of loan_amnt column
df_lendingcase.boxplot(column='loan_amnt')
```

```python
#removing outliers in loan_amnt column
df_outlier_removed=outlier_IQR_method_remove(df_lendingcase.loan_amnt)
df_outlier_removed=pd.DataFrame(df_outlier_removed)
ind_diff=df_lendingcase.index.difference(df_outlier_removed.index)

for i in range(0, len(ind_diff),1):
    df_final_lendingcase=df_lendingcase.drop([ind_diff[i]])
    df_lendingcase=df_final_lendingcase
```
